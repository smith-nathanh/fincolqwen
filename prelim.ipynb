```python
import base64
import os
from google.colab import userdata
from byaldi import RAGMultiModalModel
from claudette import models, Chat
```

```python
os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')
os.environ["ANTHROPIC_API_KEY"] = userdata.get("ANTHROPIC_API_KEY")
```

```python
RAG = RAGMultiModalModel.from_pretrained("vidore/colqwen2-v1.0-merged", verbose=1)
```

```python
RAG.index(
    input_path="drive/MyDrive/Colab_Notebooks/DLWT/project/papers/bls/",
    index_name="bls",
    store_collection_with_index=True,
    overwrite=True
)
```

```python
query = "What was the CPI percent change for recreation for urban customers in Oct 2024 excluding food and energy?"
```

```python
# retrieve the document most relevant to the query
results = RAG.search(query, k=5)
results
```

```python
# claude expect bytes
image_bytes = base64.b64decode(results[0].base64)
```
